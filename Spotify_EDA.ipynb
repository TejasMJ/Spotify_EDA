#### Based on the publicly available Spotify track metadata and audio features, the meaning of each column in the dataset is as follows:

- **track_id**: Unique identifier of the Spotify track. (Text)
- **track_name**: Name of the song. (Text)
- **track_artist**: Name of the artist who performed the song. (Text)
- **track_popularity**: Popularity score of the track ranging from 0 to 100, where higher values indicate greater popularity. (Numeric)
- **track_album_id**: Unique identifier of the album containing the track. (Text)
- **track_album_name**: Name of the album to which the track belongs. (Text)
- **track_album_release_date**: The release date of the album. (Date)
- **playlist_name**: Name of the playlist in which the track appears. (Text)
- **playlist_id**: Unique identifier of the playlist. (Text)
- **playlist_genre**: Genre category assigned to the playlist. (Text)
- **playlist_subgenre**: Subgenre classification of the playlist. (Text)
- **danceability**: A measure (0.0 to 1.0) describing how suitable a track is for dancing based on tempo, rhythm stability, beat strength, and overall regularity. Higher values indicate greater danceability. (Numeric)
- **energy**: A measure (0.0 to 1.0) representing the intensity and activity level of a track. Energetic tracks typically feel fast, loud, and dynamic. (Numeric)
- **key**: The estimated musical key of the track represented as an integer corresponding to pitch class notation (e.g., 0 = C, 1 = Câ™¯/Dâ™­, etc.). A value of -1 indicates no key was detected. (Numeric)
- **loudness**: The overall loudness of the track measured in decibels (dB). Values typically range between -60 dB and 0 dB. (Numeric)
- **mode**: Indicates whether the track is in a major or minor key. Major is represented by 1 and minor by 0. (Numeric)
- **speechiness**: A measure (0.0 to 1.0) indicating the presence of spoken words in a track. Higher values suggest more spoken-word content. (Numeric)
- **acousticness**: A confidence measure (0.0 to 1.0) indicating whether the track is acoustic. Higher values represent a higher likelihood that the track is acoustic. (Numeric)
- **instrumentalness**: A measure (0.0 to 1.0) predicting whether a track contains no vocals. Higher values indicate a greater probability that the track is instrumental. (Numeric)
- **liveness**: A measure (0.0 to 1.0) detecting the presence of an audience in the recording. Higher values indicate a higher probability that the track was performed live. (Numeric)
- **valence**: A measure (0.0 to 1.0) describing the musical positivity conveyed by the track. Higher values indicate more positive or happy-sounding tracks. (Numeric)
- **tempo**: The estimated tempo of the track in beats per minute (BPM), representing the speed or pace of the song. (Numeric)
- **duration_ms**: Duration of the track in milliseconds. (Numeric)

## Data Cleaning

```python
# Columns to fill
cols_to_fill = ['track_artist', 'track_name', 'track_album_name']

# Replace NaN with 'Unknown'
df[cols_to_fill] = df[cols_to_fill].fillna('Unknown')
```

```python
# Convert the release date column to datetime
df['track_album_release_date'] = pd.to_datetime(df['track_album_release_date'], errors='coerce')
```

```python
# Convert string/object columns to categorical where appropriate
categorical_cols = [
    'track_artist', 
    'track_name', 
    'track_album_name', 
    'playlist_name', 
    'playlist_subgenre', 
    'playlist_genre'
]

for col in categorical_cols:
    df[col] = df[col].astype('category')

# Check the updated data types
print(df.dtypes)
```
